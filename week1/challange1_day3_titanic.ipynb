```
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
df = pd.read_csv('https://raw.githubusercontent.com/dataworkshop/webinar-titanic/master/input/train.csv')
df.head()
```

```
feats = ['Pclass', 'Fare'] # lista cech do analizy
X = df[ feats ].values
y = df[ 'Survived' ] # tutaj jest nasza odpowiedź - przeżył lub nie

X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3) # dzielimy dane na treningowe i testowe

X_train.shape, X_test.shape # sprawdzamy rozmiar
```

```
# budujemy model

model = DecisionTreeClassifier(max_depth=20)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy_score(y_test, y_pred)
```

```
# budujemy model2

model = RandomForestClassifier(max_depth=10, n_estimators=50) # n_estimators to ilość drzew które zostaną zbudowane
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy_score(y_test, y_pred)
```

```
# badamy wpływ płci, jednak płeć jest zmienną kategorialną, dlatego trzeba to przerobić

df['Sex'].value_counts()
```

```
# tutaj przerabiamy płeć na wartości liczbowe

df['Sex_cat'] = df['Sex'].factorize()[0]
```

```
feats = ['Pclass', 'Fare', 'Sex_cat'] # lista cech do analizy
X = df[ feats ].values
y = df[ 'Survived' ] # tutaj jest nasza odpowiedź - przeżył lub nie

X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.3) # dzielimy dane na treningowe i testowe

X_train.shape, X_test.shape # sprawdzamy rozmiar

model = DecisionTreeClassifier(max_depth=10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy_score(y_test, y_pred)
```
